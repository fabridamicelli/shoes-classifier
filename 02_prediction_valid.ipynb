```python
import timm
from fastai.vision.all import *
from fastai.vision.widgets import *
```

```python
path = Path(os.environ["PROJECTS"])/"shoes-classifier"
(path/"data").ls()
```

```python
dls = ImageDataLoaders.from_folder(path=path/"data", train="train", valid="test",)
```

```python
dls.show_batch(max_n=8, nrows=1)
```

```python
learn = load_learner("models/best_model.pkl")
learn.dls = dls
```

```python
interp = ClassificationInterpretation.from_learner(learn,dl=learn.dls.valid)
```

```python
interp.plot_confusion_matrix()
```

```python
interp.plot_top_losses(12, nrows=2)
```
